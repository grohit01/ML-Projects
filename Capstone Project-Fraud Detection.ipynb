# IQR Method to Handle Outlier.

```python
# Handle 'No_Payments' Outlier
```

```python
# Find Q1 and Q3.
```

```python
q1, q3= np.percentile(Data['No_Payments'],[25,75]) 
```

```python
q1, q3
```

```python
# Find IQR
```

```python
iqr = q3 - q1
```

```python
iqr
```

```python
# Upper & Lower Bound
```

```python
lower_bound = q1 -(1.5 * iqr) 
upper_bound = q3 +(1.5 * iqr)
```

```python
lower_bound
```

```python
upper_bound
```

```python
New_Data=Data[(Data['No_Payments'] > upper_bound) | (Data['No_Payments'] < lower_bound)]
```

```python
# Capping.
```

```python
New_Data_Cap = Data.copy()

New_Data_Cap['No_Payments']=np.where(
    New_Data_Cap['No_Payments']>upper_bound, 
    upper_bound,
    np.where(
        New_Data_Cap['No_Payments']<lower_bound,
        lower_bound,
        New_Data_Cap['No_Payments']
    )  
)
```
